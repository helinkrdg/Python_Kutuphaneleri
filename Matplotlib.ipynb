# 1. Matplotlib Nedir?

## Matplotlib'in tanımı ve amacı

Matplotlib, Python programlama dilinde kullanılan popüler bir veri görselleştirme kütüphanesidir. Amacı, veri analizi ve görselleştirme süreçlerinde grafikler, tablolar, histogramlar, pasta grafikleri ve diğer görsel öğelerin oluşturulmasını sağlamaktır.

 Bu sayede veri bilimciler, analistler ve araştırmacılar, veri setlerini keşfetmek, trendleri belirlemek, ilişkileri gözlemlemek ve sonuçları sunmak için grafikler oluşturabilirler. 

## Python'da popüler bir veri görselleştirme kütüphanesi olarak neden kullanılır?

- **Kolay Kullanım:** Matplotlib, Python programlama diline entegre olarak çalışır ve basit bir API (Uygulama Programlama Arayüzü) sunar. Bu da kullanıcıların kolaylıkla grafikler oluşturmasını ve özelleştirmesini sağlar. Ayrıca, geniş bir dokümantasyon ve kullanıcı topluluğu mevcuttur, bu da yeni başlayanlar için yardımcı kaynaklara kolay erişim sağlar.

- **Çeşitli Grafik Türleri:** Matplotlib, birçok farklı grafik türünü destekler. Çizgi grafikleri, nokta grafikleri, bar grafikleri, pasta grafikleri, histogramlar, yüzey grafikleri gibi çeşitli grafikler oluşturabilirsiniz. Bu, farklı veri türlerini ve analiz gereksinimlerini karşılamak için esneklik sağlar.

- **Özelleştirme Seçenekleri:** Matplotlib, grafiklerinizi özelleştirmek için geniş bir yelpazede seçenek sunar. Renkler, çizgi stilleri, eksenlerin ölçekleri, etiketler, başlıklar ve daha fazlasını kişiselleştirebilirsiniz. Bu şekilde, grafiklerinizi verilerinizin ve sunumunuzun gereksinimlerine uygun hale getirebilirsiniz.

- **Yüksek Kaliteli Çıktılar:** Matplotlib, yüksek kaliteli grafiklerin oluşturulmasını sağlar. Grafiklerinizi çeşitli dosya formatlarında (PNG, JPEG, PDF, SVG) kaydedebilir veya doğrudan ekrana görüntüleyebilirsiniz. Bu, grafiklerinizi raporlar, sunumlar, makaleler veya web siteleri gibi farklı ortamlarda kullanmanızı sağlar.

- **Veri Analitiği ve Keşif için Araçlar:** Matplotlib, veri analitiği ve keşif süreçlerinde güçlü bir araçtır. Grafikler kullanarak verileri keşfedebilir, trendleri gözlemleyebilir, ilişkileri analiz edebilir ve sonuçları etkileyici bir şekilde sunabilirsiniz. Bu, veri odaklı kararlar almak ve verileri daha iyi anlamak için önemli bir yardımcıdır.

# 2. Temel Grafik Türleri

## Çizgi grafikleri

Çizgi grafikleri, verileri noktalarla birleştirerek bir çizgi şeklinde görselleştirmek için kullanılan grafik türüdür. Çizgi grafikleri, sürekli bir veri setinin değişimini veya eğilimini göstermek için yaygın olarak kullanılır. Örneğin, zamana bağlı olarak bir sürecin performansını veya bir hisse senedinin fiyat değişimini göstermek için çizgi grafikleri kullanılabilir.

```python
import matplotlib.pyplot as plt

# Veri seti
x = [1, 2, 3, 4, 5]  # x eksenindeki değerler
y = [2, 4, 6, 8, 10]  # y eksenindeki değerler

# Çizgi grafik oluşturma
plt.plot(x, y)

# Grafik başlığı ve eksen etiketleri
plt.title("Örnek Çizgi Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafik gösterimi
plt.show()
```

Bu örnekte, **x** ve **y** listeleri, çizgi grafiğindeki noktaların konumunu belirlemek için kullanılır. **plt.plot()** fonksiyonu, **x** ve **y** değerlerini alarak çizgi grafiğini oluşturur. **plt.title()**, **plt.xlabel()** ve **plt.ylabel()** fonksiyonları ile grafik başlığı ve eksen etiketleri belirlenir. Son olarak, **plt.show()** fonksiyonu ile grafiğin görüntülenmesi sağlarız.

## Nokta grafikleri

Nokta grafikleri, verileri noktalarla göstermek için kullanılan bir grafik türüdür. Her bir nokta, x ve y koordinatlarına sahip bir veri noktasını temsil eder. Nokta grafikleri, veri noktalarının dağılımını, yoğunluğunu veya ilişkisini görselleştirmek için yaygın olarak kullanılır. Nokta grafik oluşturmak için **plt.scatter(x, y)** fonksiyonu kullanılır.

```python
import matplotlib.pyplot as plt

# Veri seti
x = [1, 2, 3, 4, 5]  # x eksenindeki değerler
y = [2, 4, 6, 8, 10]  # y eksenindeki değerler

# Nokta grafik oluşturma
plt.scatter(x, y)

# Grafik başlığı ve eksen etiketleri
plt.title("Örnek Nokta Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafik gösterimi
plt.show()
```

Bu örnekte, **x** ve **y** listeleri, nokta grafiğindeki noktaların konumunu belirlemek için kullanılır. **plt.scatter()** fonksiyonu, **x** ve **y** değerlerini alarak nokta grafiğini oluşturur.

## Çizgi ve nokta grafiklerini birleştirme

```python
import matplotlib.pyplot as plt

# Veri seti
x = [1, 2, 3, 4, 5]  # x eksenindeki değerler
y = [2, 4, 6, 8, 10]  # y eksenindeki değerler

# Çizgi grafiği oluşturma
plt.plot(x, y, 'b-')

# Nokta grafiği oluşturma
plt.scatter(x, y, color='r', marker='o')

# Grafik başlığı ve eksen etiketleri
plt.title("Çizgi ve Nokta Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafik gösterimi
plt.show()
```

**plt.plot()** fonksiyonu ile çizgi grafiği oluşturulurken **'b-'** parametresi kullanılarak çizginin mavi renkte **('b')** ve sürekli **('-')** olmasını sağlıyoruz. **plt.scatter()** fonksiyonu ile nokta grafiği oluşturulurken **color='r'** parametresi kullanılarak noktaların kırmızı renkte **('r')** olmasını, **marker='o'** parametresi ile noktaların yuvarlak **('o')** şeklinde gösterilmesini sağlıyoruz.

## Bar grafikleri

Bar grafikleri, kategorik verilerin veya gruplar arasındaki karşılaştırmaların görselleştirilmesi için kullanılan grafik türüdür. Bar grafikleri, her bir kategori veya grup için dikey veya yatay çubuklar kullanarak verilerin görsel bir temsilini sağlar.

Bar grafikleri, aşağıdaki durumlarda kullanılabilir:
- Kategorik verilerin dağılımını göstermek
- Farklı grupların değerlerini karşılaştırmak
- Zaman içindeki değişimi veya trendleri göstermek
- Anket sonuçları, pazar araştırmaları veya istatistiksel verileri görselleştirmek

```python
import matplotlib.pyplot as plt

# Veri seti
kategoriler = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4', 'Kategori 5']
veriler = [20, 35, 30, 25, 15]

# Bar grafiği oluşturma
plt.bar(kategoriler, veriler)

# Grafik başlığı ve eksen etiketleri
plt.title("Örnek Bar Grafiği")
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")

# Grafik gösterimi
plt.show()
```

**Kategoriler** listesi, her bir çubuğun adını temsil eden kategori isimlerini içerir. **veriler** listesi ise her bir kategorinin değerini belirtir. **plt.bar()** fonksiyonu, kategoriler ve veriler listelerini alarak bar grafiğini oluşturur.

## Histogramlar

Histogramlar, veri setinin dağılımını görselleştirmek için kullanılan bir grafik türüdür. Histogramlar, veri değerlerinin frekanslarını veya olasılık yoğunluklarını bir aralık veya bin üzerinde gösterir. Bu grafikler, veri setinin merkezi eğilimini, dağılımını, simetrisini veya asimetrisini anlamak için kullanılır.

```python
import matplotlib.pyplot as plt

# Veri seti
veriler = [15, 20, 23, 30, 32, 35, 38, 40, 42, 45, 50, 55, 58, 60, 65, 70, 75, 80, 85]

# Histogram oluşturma
plt.hist(veriler, bins=5, edgecolor='black')

# Grafik başlığı ve eksen etiketleri
plt.title("Örnek Histogram")
plt.xlabel("Değer Aralıkları")
plt.ylabel("Frekans")

# Grafik gösterimi
plt.show()
```

**veriler** listesi, histogramın oluşturulacağı veri setini temsil eder. **plt.hist()** fonksiyonu, veriler listesini ve **bins** parametresini alarak histogramı oluşturur. bins, verilerin gruplandırılacağı aralık sayısını veya bin sayısını belirtir.

Histogramda, veriler belirtilen bin sayısıyla gruplandırılarak her bir grup için frekanslar gösterilir. Bu sayede veri setinin dağılımı ve frekansları hakkında bir anlayış elde edebilirsiniz.

## Pasta grafikleri

Pasta grafikleri (Pie charts), bir bütünün bileşenlerinin oranlarını görselleştirmek için kullanılan grafik türüdür. Pasta grafikleri, bir dairenin dilimlerini kullanarak verilerin yüzde oranlarını temsil eder. Her bir dilim, bileşenlerin göreli büyüklüğünü gösterir.

```python
import matplotlib.pyplot as plt

# Veri seti
kategoriler = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4', 'Kategori 5']
veriler = [30, 20, 15, 10, 25]

# Pasta grafiği oluşturma
plt.pie(veriler, labels=kategoriler, autopct='%1.1f%%')

# Grafik başlığı
plt.title("Örnek Pasta Grafiği")

# Grafik gösterimi
plt.show()
```

**kategoriler** listesi, pasta grafiği dilimlerinin adlarını temsil eder. **veriler** listesi ise her bir dilimin büyüklüğünü belirtir. **plt.pie()** fonksiyonu, veriler listesini alarak pasta grafiğini oluşturur. **labels** parametresi, dilimlerin etiketlerini belirtir. **autopct='%1.1f%%'** parametresi, dilimlerin yüzde değerlerini gösterir.

## Alan grafikleri

Alan grafikleri (Area charts), zaman içindeki değişimleri veya farklı kategorilerin değerlerini görselleştirmek için kullanılan grafik türüdür. Alan grafikleri, veri değerlerinin bir eksen boyunca çizgiyle birleştirilerek alanın altını doldurarak gösterilir. Bu grafikler, bir veri setinin eğilimini, dağılımını veya karşılaştırmalarını göstermek için kullanılır.

```python
import matplotlib.pyplot as plt

# Zaman serisi ve değerler
zamanlar = [1, 2, 3, 4, 5, 6, 7]
degerler = [10, 20, 15, 25, 30, 18, 12]

# Alan grafiği oluşturma
plt.fill_between(zamanlar, degerler, alpha=0.5)

# Grafik başlığı ve eksen etiketleri
plt.title("Örnek Alan Grafiği")
plt.xlabel("Zaman")
plt.ylabel("Değerler")

# Grafik gösterimi
plt.show()
```

**zamanlar** listesi, zaman serisinin değerlerini temsil eder. **degerler** listesi ise her bir zaman noktasına karşılık gelen değerleri belirtir. **plt.fill_between()** fonksiyonu, zamanlar ve degerler listelerini alarak alan grafiğini oluşturur. **alpha** parametresi, alanın dolgunluğunu ayarlar.

## Eksenlerin ölçeği ve sınırlarının ayarlanması

Eksenlerin ölçeği ve sınırlarının ayarlanması, grafiklerin odaklandığı veri aralığını belirlemek, verileri daha iyi karşılaştırmak ve grafiklerin daha anlaşılır olmasını sağlamak için kullanılır.

- Eksenlerin sınırlarını belirlemek için xlim() ve ylim() fonksiyonlarını kullanın. Bu fonksiyonlar, her bir eksene başlangıç ve bitiş noktalarını belirleyerek görüntülenecek veri aralığını sınırlar.

```python
import matplotlib.pyplot as plt

plt.xlim(0, 10)  # X ekseninin sınırları [0, 10] olarak ayarlanır
plt.ylim(0, 100)  # Y ekseninin sınırları [0, 100] olarak ayarlanır

plt.show()
```

- Eksenlerin ölçeğini belirlemek için xscale() ve yscale() fonksiyonlarını kullanın. Bu fonksiyonlar, her bir eksene ölçeklendirme yöntemlerini uygulayarak verilerin nasıl görüntüleneceğini kontrol eder.

```python
import matplotlib.pyplot as plt

plt.xscale("log")  # X ekseninin logaritmik ölçeğini ayarlar
plt.yscale("linear")  # Y ekseninin lineer ölçeğini ayarlar

plt.show()
```

Eksenlerin etiketlenmesi, grafiklerin daha anlaşılır ve okunabilir olmasını sağlar. Eksenlere metin etiketleri eklemek için aşağıdaki adımları izleyebilirsiniz:

## Eksenlerin biçimlendirilmesi

Eksenlerin biçimlendirilmesi, eksenlerin görünümünü ve formatını özelleştirmenizi sağlar.

- Eksenlerin biçimini değiştirmek için **tick_params()** fonksiyonunu kullanın. Bu fonksiyon, eksen etiketleri, işaretleyiciler ve diğer özelliklerle ilgili ayarlamalar yapmanızı sağlar.

```python
import matplotlib.pyplot as plt

plt.tick_params(axis="x", labelsize=10, rotation=45)  # X ekseninin etiket boyutunu 10, dönüş açısını 45 derece olarak ayarlar
plt.tick_params(axis="y", labelcolor="red")  # Y ekseninin etiket rengini kırmızı olarak ayarlar

plt.show()
```

- Eksen etiketlerinin formatını belirlemek için **ScalarFormatter** veya **FuncFormatter** sınıflarını kullanabilirsiniz. **ScalarFormatter** sayıların standart formatını kullanırken, **FuncFormatter** özel bir formatlama işlevini uygulamanıza izin verir.

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter

# X ekseni için ScalarFormatter kullanma
x_formatter = ScalarFormatter(useMathText=True)
x_formatter.set_powerlimits((-3, 3))  # Büyük veya küçük sayıları bilimsel gösterimden çıkarır
plt.gca().xaxis.set_major_formatter(x_formatter)

# Y ekseni için özel formatlama işlevi kullanma
def y_formatter(value, pos):
    return f"${value:.2f}"  # Sayıyı 2 ondalık basamağa kadar ve $ işaretiyle formatlar

y_formatter = FuncFormatter(y_formatter)
plt.gca().yaxis.set_major_formatter(y_formatter)

plt.show()
```

## İkincil eksenlerin kullanımı

- İkincil ekseni eklemek için **twinx()** veya **twiny()** fonksiyonlarını kullanın. **twinx()** fonksiyonu, mevcut grafiğin y ekseni üzerine ikincil bir y ekseni eklerken, **twiny()** fonksiyonu ikincil bir x ekseni ekler.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)  # x değerleri için aralık oluştur
y1 = np.sin(x)  # Birincil eksene ait veri
y2 = np.cos(x)  # İkincil eksene ait veri

fig, ax1 = plt.subplots()  # Birincil ekseni oluşturur
ax2 = ax1.twinx()  # İkincil bir y ekseni ekler

# İlk eksenin verilerini çizin
ax1.plot(x, y1, color="red", label="Birincil Eksen Veri")

# İkinci eksenin verilerini çizin
ax2.plot(x, y2, color="blue", label="İkincil Eksen Veri")

# Eksenlerin etiketlerini ve açıklamalarını belirleyin
ax1.set_xlabel("X Ekseni")
ax1.set_ylabel("Birincil Eksen Veri", color="red")
ax2.set_ylabel("İkincil Eksen Veri", color="blue")

# Eksenlerin başlıklarını ve çizgilerini belirleyin
ax1.set_title("Birincil ve İkincil Eksenler")
ax1.grid(True)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.show()

```

- İkincil ekseni özelleştirmek için ilgili eksen nesnesini kullanabilirsiniz. İkincil eksene ait etiketleri, renkleri, ölçekleri ve diğer özellikleri belirleyebilirsiniz.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)  # x değerleri için aralık oluştur
y1 = np.sin(x)  # Birincil eksene ait veri
y2 = np.cos(x)  # İkincil eksene ait veri

fig, ax1 = plt.subplots()  # Birincil ekseni oluşturur
ax2 = ax1.twinx()  # İkincil bir y ekseni ekler

# İlk eksenin verilerini çizin
ax1.plot(x, y1, color="red", label="Birincil Eksen Veri")

# İkinci eksenin verilerini çizin
ax2.plot(x, y2, color="blue", label="İkincil Eksen Veri")

# Eksenlerin etiketlerini ve açıklamalarını belirleyin
ax1.set_xlabel("X Ekseni")
ax1.set_ylabel("Birincil Eksen Veri", color="red")
ax2.set_ylabel("İkincil Eksen Veri", color="blue")

# Eksenlerin başlıklarını ve çizgilerini belirleyin
ax1.set_title("Birincil ve İkincil Eksenler")
ax1.grid(True)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.show()
```

### Şeffaflık:

Matplotlib'de şeffaflık değeri 0 ile 1 arasında bir değer alır, 0 tamamen şeffafken 1 tamamen opaktır. Şeffaflık ayarlamak için alpha parametresini kullanabilirsiniz.

```python
plt.scatter(x, y, alpha=0.5)  # Şeffaflığı 0.5 olan bir nokta grafiği
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color="blue", label="Sin(x)")
plt.plot(x, y2, color="red", label="Cos(x)")

plt.fill_between(x, y1, y2, where=(y1 > y2), color="green", alpha=0.3)
plt.fill_between(x, y1, y2, where=(y1 <= y2), color="yellow", alpha=0.3)

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Sin(x) ve Cos(x) Grafikleri")
plt.legend()

plt.show()
```

Bu örnekte, **fill_between()** fonksiyonunu kullanarak **sin(x)** ve **cos(x)** fonksiyonlarının arasını dolgu ile renklendirdik. **where** parametresi ile belirli bir koşulu sağlayan bölgeleri dolgu olarak belirledik. **Sin(x) > Cos(x)** olduğu durumda yeşil, **Sin(x) <= Cos(x)** olduğu durumda sarı dolgu kullanıldı. **color** parametresi ile dolgu rengini, **alpha** parametresi ile şeffaflık değerini belirledik.

# 5. Eklemeler ve Anahatlar

## Metin eklemek

Matplotlib'de metin eklemek için plt.text() fonksiyonunu kullanabilirsiniz. Bu fonksiyon, koordinatları ve metni parametre olarak alır ve belirtilen koordinatlarda metni grafik üzerine ekler.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.text(3, 6, "Orta Nokta", fontsize=12, ha="center")
plt.text(5, 10, "Son Nokta", fontsize=12, ha="right")

plt.show()
```

YBu örnekte, **plt.text()** fonksiyonuyla **(3, 6)** koordinatlarına **"Orta Nokta"** metni, **(5, 10)** koordinatlarına da **"Son Nokta"** metni ekledik. **fontsize** parametresi ile metin boyutunu ve ha parametresi ile yatay hizalamayı belirledik.

Matplotlib'de anahatlar, klemelerden daha gelişmiş bir şekilde metin eklemek için kullanılır. **plt.annotate()** fonksiyonu kullanılarak belirli bir noktaya metin eklenirken, oklar, çizgiler veya şekillerle bağlantılar yapılabilir.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.annotate("Orta Nokta", xy=(3, 6), xytext=(4, 8), arrowprops=dict(arrowstyle="->"))

plt.show()
```

Bu örnekte, **plt.annotate()** fonksiyonuyla **(3, 6)** koordinatlarına **"Orta Nokta"** metni ekledik. **xytext** parametresi ile metnin konumunu, **arrowprops** parametresi ile ok stilini belirledik.

## Oklar ve anahatlar çizmek

Oklar ve anahatlar çizmek için **plt.arrow()** ve **plt.annotate()** fonksiyonlarını kullanabilirsiniz.

- **plt.arrow() Fonksiyonu**

**plt.arrow()** fonksiyonu, belirli bir noktadan başlayarak bir ok çizmek için kullanılır. Bu fonksiyon, başlangıç noktası **(x, y)**, okun yönü **(dx, dy)**, okun uzunluğu **(length)** ve diğer özelleştirme parametrelerini alır.

```python
import matplotlib.pyplot as plt

plt.arrow(0, 0, 0.5, 0.5, head_width=0.1, head_length=0.1, fc='blue', ec='black')

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True)
plt.show()
```

Bu örnekte, **(0, 0)** noktasından başlayarak **(0.5, 0.5)** yönde bir ok çizdik. **head_width** ve **head_length** parametreleri ile ok başının boyutunu belirledik. **fc** ve **ec** parametreleri ile okun dolgu rengi ve kenar rengini belirttik.

- **plt.annotate() Fonksiyonu**

Bir noktaya metinle birlikte bir ok veya çizgi bağlantısı eklemek için kullanılır. Bu fonksiyon, metin (text), noktanın konumu **(xy)**, metnin konumu **(xytext)** ve diğer özelleştirme parametrelerini alır.

```python
import matplotlib.pyplot as plt

plt.annotate('Nokta A', xy=(0.5, 0.5), xytext=(0.7, 0.7),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

Bu örnekte, **(0.5, 0.5)** noktasında **"Nokta A"** metniyle bir ok çizdik. **xy** parametresi okun hedef noktasını belirtirken, xytext parametresi metnin konumunu belirtir. **arrowprops** parametresi ile okun özelliklerini belirttik, burada okun rengi **(facecolor)** ve stili **(arrowstyle)** belirlendi.

## Raster ve vektör grafik formatlarında kaydetmek

Oluşturduğunuz grafikleri raster (örneğin PNG, JPEG) veya vektör (örneğin SVG, PDF) formatlarda kaydetmek oldukça kolaydır. Matplotlib, çıktı formatını belirlemenizi sağlayan **savefig()** fonksiyonunu sağlar.

**1. Raster Formatlar (PNG, JPEG, vs.):**

```python
import matplotlib.pyplot as plt

# Grafik çizimi
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

# Grafik kaydetme
plt.savefig("grafik.png")  # PNG formatında kaydetme
plt.savefig("grafik.jpg")  # JPEG formatında kaydetme
```

```python
import matplotlib.pyplot as plt

# Grafik çizimi
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

# Grafik kaydetme
plt.savefig("grafik.svg")  # SVG formatında kaydetme
plt.savefig("grafik.pdf")  # PDF formatında kaydetme
```

**savefig()** fonksiyonuna çeşitli parametreler ekleyebiliriz. Örneğin, DPI (dot per inch) değerini ayarlamak için **dpi** parametresini kullanabiliriz. Ayrıca, kaydedilen grafikteki beyaz alanları kırpmak için bbox_inches parametresine "tight" değerini verebiliriz..

Seaborn kütüphanesinde bulunan tips veri seti kullanarak 2 satır ve 2 sütunlu bir alt grafik yapısı oluşturalım.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# tips veri setini yükleme
tips = sns.load_dataset('tips')

# İki satır ve iki sütunlu bir alt grafik yapısı oluşturma
plt.figure(figsize=(10, 8))

# 1. alt grafik: Total Bill histogramı
plt.subplot(2, 2, 1)
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Total Bill')

# 2. alt grafik: Tip miktarı histogramı
plt.subplot(2, 2, 2)
sns.histplot(data=tips, x='tip', kde=True)
plt.title('Tip')

# 3. alt grafik: Yemek türü dağılımı
plt.subplot(2, 2, 3)
sns.countplot(data=tips, x='day')
plt.title('Day')

# 4. alt grafik: Cinsiyet dağılımı
plt.subplot(2, 2, 4)
sns.countplot(data=tips, x='sex')
plt.title('Sex')

# Grafikleri gösterme
plt.tight_layout()
plt.show()
```

## Eksenlerin paylaşılması ve senkronizasyon

Eksenlerin paylaşılması ve senkronizasyonu, matplotlib kütüphanesinde **sharex** ve **sharey** parametrelerini kullanarak gerçekleştirebiliriz. Bu parametreler, birden çok alt grafik arasında x veya y eksenlerini paylaşmayı sağlar.

# 7. 3D Grafikler

3D grafikler, verilerin üç boyutlu uzayda daha detaylı bir şekilde görselleştirilmesini sağlar. Bu, verilerin derinlik, perspektif ve ilişkilerini daha iyi anlamamıza olanak tanır. 3D grafikler, veri analizi sürecinde daha kapsamlı ve etkileşimli bir keşif sağlar ve bazen verilerdeki desenleri ve trendleri daha net bir şekilde ortaya koyar.

## 3D çizgi grafikleri

3D çizgi grafikleri, üç boyutlu uzayda verileri görselleştirmek için kullanılan grafik türlerindendir. Her bir veri noktası üç boyutlu bir koordinat sistemde temsil edilir ve çizgiler bu noktaları birleştirerek bir modelin veya veri setinin 3D görüntüsünü oluşturur.

```python
import numpy as np
import matplotlib.pyplot as plt

# Veri oluşturma
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# 3D çizim için figür oluşturma
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Çizgi grafiğini çizme
ax.plot(x, y, z)

# Eksen etiketlerini ekleme
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Grafik gösterme
plt.show()
```

Bu örnekte, x, y ve z koordinatlarındaki veriler kullanılarak bir 3D çizgi grafiğini çizdik. **projection='3d'** parametresiyle figürün 3D görünümü ayarladık. Ardından plot fonksiyonu kullanılarak çizgi grafiğini oluşturduk.

## 3D nokta grafikleri

3D nokta grafikleri, üç boyutlu uzayda verilerin nokta olarak görselleştirildiği grafik türleridir. Her bir veri noktası, üç boyutlu bir koordinat sistemde belirli bir konumu temsil eder.

```python
import numpy as np
import matplotlib.pyplot as plt

# Veri oluşturma
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 3D çizim için figür oluşturma
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Nokta grafiğini çizme
ax.scatter(x, y, z)

# Eksen etiketlerini ekleme
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Grafik gösterme
plt.show()
```

Bu örnekte, x, y ve z koordinatlarındaki rastgele veriler kullanılarak 3D nokta grafiği çizidik. **projection='3d'** parametresiyle figürün 3D görünümü ayarladık. **scatter** fonksiyonu kullanılarak noktaları belirledik. Son olarak, her bir eksenin etiketleri eklenmiş ve grafiği görüntüledik.

## 3D yüzey grafikleri

3D yüzey grafikleri, üç boyutlu uzayda verilerin yüzey olarak görselleştirildiği grafik türleridir. Veri noktaları, bir ızgara üzerindeki noktaların konumlarına dayalı olarak yüzeyi oluşturur.

```python
import numpy as np
import matplotlib.pyplot as plt

# Veri oluşturma
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D çizim için figür oluşturma
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Yüzey grafiğini çizme
ax.plot_surface(X, Y, Z, cmap='viridis')

# Eksen etiketlerini ekleme
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Grafik gösterme
plt.show()
```

Bu örnekte, X ve Y koordinatlarının ızgara düzeninde oluşturulduğu ve Z değerlerinin bu ızgara üzerindeki noktalara bağlı olarak hesaplandığı bir veri seti kullanılmıştır. plot_surface fonksiyonu kullanılarak 3D yüzey grafiği çizdik. **cmap='viridis'** parametresi, renk paletini belirleriz.

## Özel grafik nesneleri oluşturma

Özel grafik nesneleri oluşturabiliriz. Bu nesneler, grafiklerin farklı bileşenlerini veya özel öğelerini temsil eder. Örneğin, çizgi, metin, daire veya çokgen gibi özel şekiller oluşturabiliriz.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Boş bir figür ve eksen oluştur
fig, ax = plt.subplots()

# Çokgenin köşe noktalarını belirle
vertices = [(0, 0), (1, 0), (0.5, 0.7), (0, 0)]

# Çokgen nesnesini oluştur
polygon = patches.Polygon(vertices, closed=True, edgecolor='red', facecolor='yellow')

# Çokgeni eksen üzerine ekle
ax.add_patch(polygon)

# Eksenlerin sınırlarını ayarla
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.2)

# Grafik gösterme
plt.show()
```

Bu örnekte, **patches.Polygon** sınıfını kullanarak bir çokgen nesnesi oluşturduk. **vertices** listesi ile çokgenin köşe noktalarını belirledik. Ardından, bu çokgen nesnesini eksen üzerine ekledik **(ax.add_patch)**. Çokgenin kenar rengini **(edgecolor)** ve iç dolgusunu **(facecolor)** belirledik. Son olarak, eksenlerin sınırlarını ayarlayarak ve **plt.show()** ile grafikleri gösterdik.

## Veri etiketleri ve açıklamaları eklemek

Veri etiketlerini ve açıklamalarını grafiklere eklemek için matplotlib.pyplot modülündeki **text()** fonksiyonunu kullanabilirsiniz. Bu fonksiyonu kullanarak belirli koordinatlarda metin yerleştirebilirsiniz.

```python
import matplotlib.pyplot as plt

# Veri noktaları
x = [1, 2, 3, 4, 5]
y = [3, 6, 2, 7, 5]

# Grafik oluşturma
fig, ax = plt.subplots()

# Veri noktalarını çiz
ax.plot(x, y, marker='o', linestyle='-', color='blue')

# Veri noktalarına etiket ekleme
for i in range(len(x)):
    ax.text(x[i], y[i]+0.5, f'({x[i]}, {y[i]})', ha='center')

# Eksen etiketleri
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Grafik gösterme
plt.show()
```

Bu örnekte, **text()** fonksiyonunu kullanarak her veri noktasının üstüne etiket ekliyoruz. **text()** fonksiyonu ile etiketin yerleştirileceği koordinatları ve etiketin içeriğini belirliyoruz. **ha='center'** parametresi, etiketin yatay hizalamasını merkeze almasını sağlar.

Bu şekilde veri noktalarına veya grafik üzerindeki herhangi bir konuma etiket veya açıklama ekleyebilirsiniz.
